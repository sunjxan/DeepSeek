```python
# 假设的长文本输入
long_text = """
智能环境感知系统是一个集成了多种传感器、计算平台和AI算法的先进技术，它能够实时感知周围环境的变化，并做出智能反应。系统通过数据采集、处理和反馈环节，实时监控和分析环境中的物理、化学以及生物特征，从而优化人类的居住和工作环境。

例如，这样的系统可以在室内根据空气质量、温度、湿度以及光照等数据自动调节空调、照明系统和空气净化设备，创造一个最舒适的环境。它还能够基于用户的习惯和偏好提供个性化的智能服务，如调整温度、声音、灯光等，以最大程度提高居住者的体验感。
"""

# 分段函数：将长文本按最大长度分段
def split_text(text, max_length):
    return [text[i:i+max_length] for i in range(0, len(text), max_length)]

# 分段处理文本
def process_in_chunks(text, max_length):
    chunks = split_text(text, max_length)
    responses = []
    for chunk in chunks:
        prompt = f"""
        现在，我给你一个全新的任务，要求你描述一个你从未见过或学习过的概念——“智能环境感知系统”。
        这里是你需要处理的内容：
        {chunk}
        """
        # 获取每段的模型输出
        response = chat(messages=[
            {"role": "user", "content": prompt}
        ])
        responses.append(response.strip())
    return "\n---\n".join(responses)

# 假设token的最大限制是1000个字符
max_token_length = 100

# 处理并输出最终结果
final_output = process_in_chunks(long_text, max_token_length)
print(final_output)
```

```python
# 初始化对话历史
conversation_history = []

def chat_with_memory(message):
    # 将用户消息添加到对话历史
    conversation_history.append({"role": "user", "content": message})

    # 将对话历史作为上下文传给模型
    response = chat(
        messages=conversation_history
    )

    # 将助手的回答添加到对话历史
    conversation_history.append({"role": "assistant", "content": response})

    return response

user_input_1 = "我想点一份宫保鸡丁，请问现在有吗？"
print("用户:", user_input_1)
response_1 = chat_with_memory(user_input_1)
print("助手:", response_1)

# 第二次用户提问
user_input_2 = "不好意思，我刚刚点的什么？"
print("\n用户:", user_input_2)
response_2 = chat_with_memory(user_input_2)
print("助手:", response_2)
```

```python
propose_prompt = """
输入: 2 8 8 14
可能的下一步行动：
2 + 8 = 10 (left: 8 10 14)
8 / 2 = 4 (left: 4 8 14)
14 + 2 = 16 (left: 8 8 16)
2 * 8 = 16 (left: 8 14 16)
8 - 2 = 6 (left: 6 8 14)
14 - 8 = 6 (left: 2 6 8)
14 / 2 = 7 (left: 7 8 8)
14 - 2 = 12 (left: 8 8 12)
输入: {inputs}
可能的下一步行动：
"""
 
value_prompt = """
评估给定的数字是否可以达到24，输出sure(20)/likely(1)/impossible(0.001)，只输出分值
10 14
10 + 14 = 24
left: 20
11 12
11 + 12 = 23
12 - 11 = 1
11 * 12 = 132
11 / 12 = 0.91
left: 0.001
4 4 10
4 + 4 + 10 = 8 + 10 = 18
4 * 10 - 4 = 40 - 4 = 36
(10 - 4) * 4 = 6 * 4 = 24
left: 20
4 9 11
9 + 11 + 4 = 20 + 4 = 24
left: 20
5 7 8
5 + 7 + 8 = 12 + 8 = 20
(8 - 5) * 7 = 3 * 7 = 21
无法达到24, 但数字在合理范围内
left: 1
5 6 6
5 + 6 + 6 = 17
(6 - 5) * 6 = 1 * 6 = 6
无法达到24, 但数字在合理范围内
left: 1
10 10 11
10 + 10 + 11 = 31
(11 - 10) * 10 = 10
10 10 10 太大了
left: 0.001
1 3 3
1 * 3 * 3 = 9
(1 + 3) * 3 = 12
1 3 3 太小了
left: 0.001
{inputs}
"""
 
# 第一轮思考的处理函数，输入为用户提供的初始数字
def first_think(inputs):
    # 使用格式化的propose_prompt构造第一轮思考的提示，调用chatgpt函数生成提案并按行拆分
    messages = [{"role": "user", "content": propose_prompt.format(inputs=inputs)}]
    proposals = chat(messages=messages).split('\n')
    # 返回提案的索引和提案内容
    return proposals
 
# 第一轮评估函数，评估每个提案的“价值”
def evaluate(proposals):
    values = []  # 存储每个提案的评估值
    for proposal in proposals:
        # 从提案中提取出“left”部分的数字，以便后续评估
        current_numbers = proposal.strip().split('left: ')[-1].split(')')[0]
        if current_numbers in value_cache:
            # 如果当前数字已在缓存中，则跳过评估
            values.append(0)
            continue
        # 否则，调用value_prompt评估当前提案的价值，生成3个评估回复
        messages = [{"role": "user", "content": value_prompt.format(inputs=current_numbers)}]
        value_outputs = chat(messages=messages, n=3)
        value_cache.add(current_numbers)  # 将当前数字添加到缓存中
        value = 0  # 初始化价值评分
        try:
            # 尝试从3个评估回复中提取并汇总评分
            value = sum([float(output.split('\n')[-1].split('left: ')[-1]) for output in value_outputs])
        except:
            pass
        # 将评分添加到值列表中
        values.append(value)
    return values  # 返回所有提案的评估分值
 
# 第一轮筛选函数，根据评估分值选择最优的前5个提案
def pick(proposals, values):
    # 创建一个包含提案索引的列表，用于后续选择和评估
    ids = list(range(len(values)))
    # 根据分值排序选择最有可能到达目标24的前5个提案
    select_ids = sorted(ids, key=lambda i: values[i], reverse=True)[:5]
    # 根据排序结果提取相应的提案内容
    select_proposals = [proposals[select_id] for select_id in select_ids]
    return select_proposals  # 返回筛选后的前5个提案
 
# 第二轮思考函数，根据第一轮筛选出的提案生成下一步操作
def think(select_proposals):
    proposals = []  # 用于存储第二轮生成的提案
    for proposal in select_proposals:
        # 从第一轮筛选出的提案中提取当前的数字
        current_numbers = proposal.strip().split('left: ')[-1].split(')')[0]
        # 调用chatgpt生成可能的下一步操作提案
        messages = [{"role": "user", "content": propose_prompt.format(inputs=current_numbers)}]
        res = chat(messages=messages).split('\n')
        # 将新生成的提案与第一轮的提案拼接
        for item in res:
            proposals.append(proposal + ' ' + item) 
    # 返回第二轮提案的索引和内容
    return proposals
 
# 第三轮评估函数，判断提案是否成功得到24
def last_evaluate(proposals):
    result = []  # 存储符合条件的最终结果
    for proposal in proposals:
        # 提取“left”部分的数字并检查是否为24
        current_numbers = proposal.strip().split('left: ')[-1].split(')')[0]
        if current_numbers == '24':
            result.append(proposal)  # 如果数字为24，加入结果列表
    return result  # 返回得到24的提案
 
# 获取最终结果函数，输出所有符合条件的算式
def get_result(result):
    if not len(result):
        print("无法获得24")  # 如果没有符合条件的算式，输出提示信息
    else:
        print("获得24的算式有:")
        # 输出所有符合条件的算式
        for line in result:
            print(line)


# 程序主入口，执行游戏的所有步骤
value_cache = set()  # 用于缓存已评估的数字

inputs = "5 7 11 13"  # 用户输入的数字

print("第一轮：思考生成中......")
# 进行第一轮思考并获取提案
proposals = first_think(inputs)
print("第一轮思考完毕，结果：")
for proposal in proposals:
    print(proposal)
print()

print("第一轮：评估结果中......")
# 评估第一轮的提案
values = evaluate(proposals)
print("第一轮评估完毕，结果：")
for index, proposal in enumerate(proposals):
    print(proposal, "得分：", values[index])
print()

print("第一轮：筛选候选项......")
select_proposals = pick(proposals, values)
print("第一轮候选项筛选完毕，结果：")
for proposal in select_proposals:
    print(proposal)
print()

print("第二轮：思考生成中......")
# 进行第二轮思考并获取提案
proposals = think(select_proposals)
print("第二轮思考完毕，结果：")
for proposal in proposals:
    print(proposal)
print()

print("第二轮：评估结果中......")
# 评估第二轮的提案
values = evaluate(proposals)
print("第二轮评估完毕，结果：")
for index, proposal in enumerate(proposals):
    print(proposal, "得分：", values[index])
print()

print("第二轮：筛选候选项......")
select_proposals = pick(proposals, values)
print("第二轮候选项筛选完毕，结果：")
for proposal in select_proposals:
    print(proposal)
print()

print("第三轮：思考生成中......")
# 进行第三轮思考并获取提案
proposals = think(select_proposals)
print("第三轮思考完毕，结果：")
for proposal in proposals:
    print(proposal)
print()

print("第三轮：评估结果中......")
# 评估第三轮的提案
result = last_evaluate(proposals)
print("第三轮评估完毕，结果：")
get_result(result)
```
